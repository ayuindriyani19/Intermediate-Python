### 5. Get rid of the missing value:
### a. Drop columns which has missing value more than 50% of the total dat

### Menghapus data yang percentase kosong lebih dari 50

```python
data_hotel.dropna(thresh=0.5*len(data_hotel), axis=1, inplace=False)
```

### b. Fill the missing value in agent with the mean

### Mencari rata-rata column agent

```python

mean_agent =data_hotel['agent'].mean()
mean_agent
```

### Mengisi data agent yang masih kosong dengan hasil nilai rata

```python
data_hotel['agent']=data_hotel['agent'].fillna(mean_agent)
data_hotel
```

### 8. Visualize the arrival_date_month to see in what month the most visitors come

```python
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
visitors = [0]*len(months)
```

Visualize the arrival_date_month to see in what month the most visitors come

```python
visualize = hotel_clean["arrival_date_month"].value_counts()
visualize
```

```python
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'December']
visitors = [0]*len(months)
```

```python
for month in months:
    idx = months.index(month)
    visitors[idx] = visualize[month]
visitors
```

```python
plt.barh(months, visitors, color='blue')
plt.ylabel("Month", fontdict = {"fontsize": 18})
plt.xlabel("Number of Visitors", fontdict = {"fontsize": 18})
plt.show()
```

```python
data_month_name = np.array(['January','February','March','April','May','June','July','August','September','October','November','December'])
data_month_value = np.array([])
data_month_canceled =np.array([])
data_month_checkOut =np.array([])
data_month_noShow =np.array([])


for month in data_month_name:
    x = np.equal(hotel_clean['arrival_date_month'],month)
    data_month_value = np.append(data_month_value,np.sum(x))
    data_month_canceled = np.append(data_month_canceled, hotel_clean[
            (hotel_clean['reservation_status']=='Canceled')
            &(hotel_clean['arrival_date_month']==month)
            ].value_counts().sum())
    data_month_checkOut = np.append(data_month_checkOut, hotel_clean[
            (hotel_clean['reservation_status']=='Check-Out')
            &(hotel_clean['arrival_date_month']==month)
            ].value_counts().sum())
    data_month_noShow = np.append(data_month_noShow, hotel_clean[
            (hotel_clean['reservation_status']=='No-Show')
            &(hotel_clean['arrival_date_month']==month)
            ].value_counts().sum())
    
    

plt.figure(figsize=(20, 8))
plt.title("Arrival Date Month")
plt.xlabel("Month")
plt.ylabel("Total")


plt.bar(data_month_name,data_month_value, color='blue',label='Total')
plt.bar(data_month_name,data_month_checkOut,color='green',label='Check Out')
plt.bar(data_month_name,data_month_canceled,color='red',label='Canceled')
plt.bar(data_month_name,data_month_noShow,color='black',label='No-Show')


plt.legend()

plt.show()
```

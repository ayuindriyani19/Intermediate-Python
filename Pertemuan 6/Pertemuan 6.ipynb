## Start Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use("seaborn")
```

```python
df = pd.read_csv("data_sub.csv")
df.head(10)
```

```python
# Define data
## X(Age)
ages = df["Age"]

## Y(Salaries)
dev_general_salaries = df["All_Devs"]
dev_js_salaries = df["JavaScript"]
dev_py_salaries = df["Python"]
```

```python
fix, ax = plt.subplots()
```

```python
fix, ax = plt.subplots()
ax.plot(ages, dev_py_salaries, label='Python')
ax.plot(ages, dev_js_salaries, label='JavaScript')

ax.plot(ages, dev_general_salaries, color='red',
         linestyle='--', label='All Devs')

ax.legend()

ax.set_title('Median Salary (USD) by Age')
ax.set_xlabel('Ages')
ax.set_ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1) #by default it is 1x1

ax1.plot(ages, dev_general_salaries, color='red',
         linestyle='--', label='All Devs')

ax2.plot(ages, dev_py_salaries, label='Python')
ax2.plot(ages, dev_js_salaries, label='JavaScript')

# Define titles and labels
## Ax 1
ax1.legend()
ax1.set_title("general Salary For Developers (USD) by Age")
ax1.set_xlabel("Ages")
ax1.set_ylabel("Salary (USD)")

## Ax 2
ax2.legend()
ax2.set_title("general Salary For Py and Js Developers (USD) by Age")
ax2.set_xlabel("Ages")
ax2.set_ylabel("Salary (USD)")

# Output
plt.tight_layout()
plt.show()
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(3) #by default it is 1x1
plt.margins(x=0.5, y=0)

ax1.plot(ages, dev_general_salaries, color='red',
         linestyle='--', label='All Devs')

ax2.plot(ages, dev_py_salaries, "-.", color="green", label='Python')

ax3.plot(ages, dev_js_salaries, label='JavaScript')

# Define titles and labels
## Ax 1
ax1.legend()
ax1.set_title("general Salary For Developers (USD) by Age")
ax1.set_xlabel("Ages")
ax1.set_ylabel("Salary (USD)")

## Ax 2
ax2.legend()
ax2.set_title("general Salary For Py Developers (USD) by Age")
ax2.set_xlabel("Ages")
ax2.set_ylabel("Salary (USD)")

## Ax 3
ax3.legend()
ax3.set_title("general Salary For Js Developers (USD) by Age")
ax3.set_xlabel("Ages")
ax3.set_ylabel("Salary (USD)")

# Output
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

# We need to draw the canvas, otherwise the labels won't be positioned and 
# won't have values yet.
fig.canvas.draw()

labels = [item.get_text() for item in ax.get_xticklabels()]
labels[1] = "tes"

ax.set_xticklabels(labels)

plt.show()
```
